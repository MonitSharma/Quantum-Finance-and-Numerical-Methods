# Quantum Amplitude Amplification

For a classical randomized algorithm ,that have a success *probability* of $p$, repeated $j$ times, boost the 
success probability to $jp$ ( assuming $jp<< 1$). Basically, an additional run is boosting the success probability.


----

 A quantum analouge for that would be to boost the *amplitude* of being in the certain subspace of the Hilbert space.

$\mathcal{H}$ be the Hilber space. Every **boolean** function $\chi : \mathbb{Z} → \{0,1\}$ induces a partition of $\mathcal{H}$ into a direct sum of two subspaces, a **good** one and a **bad** one.

------

The **good** one is spanned by the set of the basis states $|x⟩ \in \mathcal{H}$ for which $\chi(x) = 1$ and the **bad** one is the orthogonal one.

Every pure state $|ζ⟩$ has a unique decomposition as $|ζ⟩ = |ζ_1⟩ + ||ζ_0⟩$

where one is the projection on the good state, the other is the projection on the bad state.

---

$a_{\zeta} = ⟨\zeta_1|\zeta_1⟩$ is the probability of measuring the state in a good state, similary for bad state we have $b_{\zeta}$

Let $\mathcal{A}$ be a quantum algorithm that acts in $\mathcal{H}$ and has no measurements.

The *amplification* process is realized by repeatedly applying the following:

$$ -\mathcal{A} S_0 \mathcal{A}^{-1} S_{\chi}$$


Here the operator $S_\chi$ conditionally changes the sign of the amplitudes of the good state, while the operator $S_0$ changes the sign of the amplitude iff the state is the zero state.



----

This complete operator is also called the operator $\mathcal{Q}$, so we have

$$\mathcal{Q}|Ψ_1⟩ = (1-2a)|\Psi_1⟩ - 2a |\Psi_0⟩$$

$$\mathcal{Q}|Ψ_0⟩ = 2(1-a)|\Psi_1⟩ - (1-2a) |\Psi_0⟩$$

where $a = ⟨Ψ_1|\Psi_1⟩$

This can also be realised by the operator $$U_\psi U_{\psi_0}$$

which is composed oftwo reflections.

The first operatoe $U_{\psi_0} = \mathcal{I} - \frac{2}{1-a}|\Psi_0⟩ ⟨\Psi_0|$ implements a reflection through the ray spanned by the vector $|\Psi_0⟩$ while the second one $U_{\psi} = \mathcal{I} - 2|\Psi⟩⟨\Psi|$ implements a reflection through the ray spanned by the vector $|\Psi⟩$

For the orthogonal complement $\mathcal{H}_{\psi}^{p}$ of $\mathcal{H}_{\psi}$ in $\mathcal{H}$.

The operator $\mathcal{A}S_0 \mathcal{A}^{-1}$ acts as identity on $\mathcal{H}_{\psi}^{p}$ operator $Q$ acts as $-S_\chi$ , thus $Q^2$ acts as identity, hence the eigenvector of $Q$ in $\mathcal{H}_{\psi}^{p}$ has eigenvalue of $\pm 1$

## Quantum Amplitude Estimation


Given an operator $\mathcal{A}$ that acts as 

$$ \mathcal{A}|0\rangle = \sqrt{1 - a}|\Psi_0\rangle + \sqrt{a}|\Psi_1\rangle $$

**Quantum Amplitude Estimation** is the task of finding an estimate for the amplitude $a$ of the state $|\Psi_1⟩$

$$a = |\langle\Psi_1 | \Psi_1\rangle|^2.$$




### Bernoulli Variable Example

$\mathcal{A}$ describes a Bernoulli random variable with (unknown) sucess probability $p$

$$ \mathcal{A}|0\rangle = \sqrt{1 - p}|0\rangle + \sqrt{p}|1\rangle. $$



### How to model the operator on quantum computer?

We can model this operator with a rotation around the Y-axis

$$ \mathcal{A} = R_Y(\theta_p), \theta_p = 2\sin^{-1}(\sqrt{p}). $$

----

The grover operator in this case is 

$$ \mathcal{Q} = R_Y(2\theta_p), $$

whose powers are easy to calculate

$$ \mathcal{Q}^k = R_Y(2k\theta_p) $$

````python
p = 0.8
````

The circuits $\mathcal{A}$ and $\mathcal{Q}$

````python
%pip install qiskit
````

````python
import numpy as np
from qiskit.circuit import QuantumCircuit, QuantumRegister
from qiskit.circuit.library import GroverOperator
````

````python
class BernoulliA(QuantumCircuit):
    """A circuit representing the Bernoulli A operator."""

    def __init__(self, probability):
        super().__init__(1)  # circuit on 1 qubit

        theta_p = 2 * np.arcsin(np.sqrt(probability))
        self.ry(theta_p, 0)


class BernoulliQ(QuantumCircuit):
    """A circuit representing the Bernoulli Q operator."""

    def __init__(self, probability):
        super().__init__(1)  # circuit on 1 qubit

        self._theta_p = 2 * np.arcsin(np.sqrt(probability))
        self.ry(2 * self._theta_p, 0)

    def power(self, k):
        # implement the efficient power of Q
        q_k = QuantumCircuit(1)
        q_k.ry(2 * k * self._theta_p, 0)
        return q_k
````

````python
A = BernoulliA(p)
Q = BernoulliQ(p)
````

# How to use QAE to estimate an expected value $\mathbb{E}[X]$ for a given random variable $X$



Suppose $f : \mathbb{R} → [0,1]$ and a corresponding unitary operator $F$ given by

$$ F|x⟩ |0⟩ = \sqrt{1-f(x)}|x⟩ |0⟩ + \sqrt{f(x)}|x⟩ |1⟩$$

Such an operator $F$ can be constructed.

-----


Suppose a discrete random variable $X$ taking values in $Ω_X = \{x_i\}_{i=0}^{N-1}$ , where $N= 2^n$ for a given $n$  with corresponding probabilities $p_x = \mathbb{P}[X=x_i]$. Then the expectation value can be written as 

$$ \mathbb{E}[f(x)] = \sum_{i=0}^{N-1} p_{x_i} f(x_i) $$

## Quantum Risk Analysis

**How to use amplitude estimation to approximate the expected value of a random variable**

Suppose a quantum state

$$|\psi_n⟩ = \sum_{i=0}^{N-1} \sqrt{p_i} |i⟩_n$$

The state $|i⟩_n$ is one of the $N$ possible realization of a bounded discrete random variable $X$, which for instance can represent discretized interest rates or the valie of a portfolio.


------


We consider a function $f$ and a corresponsing operator

$$ F: |i⟩_n |0⟩ → |i⟩_n \big( \sqrt{1-f(i)}|0⟩ + \sqrt{f(i)}|1⟩\big)$$

Applying $F$ to $|\psi⟩_n |0⟩$ leads to the state 

$$ \sum_{i=0}^{N-1} \sqrt{1-f(i)}|i⟩_n |0⟩ + \sum_{i=0}^{N-1} \sqrt{f(i)}\sqrt{p_i}|i⟩_n |1⟩$$


Now we can use the **Amplitude Estimation** to approximate the probability of measuring $|1⟩$ in the last qubit, which equals $\sum_{i=0}^{N-1}p_i f(i)$ which is also the $\mathbb{E}[f(X)]$






### Example : Pricing European Call Options


We have a strie price $K$ and an underlying asset whose spot price at maturity $S_T$ follows a given random distribution.

----

The corresponding payoff function is defined as:

$$ \text{max }\{S_T - K,0\}$$

In the following tutorial, a quantum algorithm based on amplitude estimation is used to estimate the expected payoff 

$$ \mathbb{E}[\text{max}\{S_T -K,0\}$$

as well ass the corresponding $Δ$ which is the derivative of the option pricing with respect to the spot price, defined as:

$$Δ = \mathbb{P}[S_T \ge K]$$

## Uncertainity Model

We construct a circuit to load a log-normal distributon into a quantum state. The distribution is truncated to a given interval $[\text{low, high}]$ and discretized using $2^n$ grid points, where $n$ denotes the number of qubits used. The unitary operator corresponding to the circuit implements the following:

$$ |0⟩_n → |\psi⟩_n = \sum_{i=0}^{2^n -1} \sqrt{p_i}|i⟩_n$$

where $p_i$ denote the probabilities corresponding to the truncated and discretized distribution and where $i$ is mapped to the right interval using the affine map.

$$ \{0, \ldots, 2^n-1\} \ni i \mapsto \frac{\text{high} - \text{low}}{2^n - 1} * i + \text{low} \in [\text{low}, \text{high}]. $$

#### The parameters that define the distribution

````python
# number of qubits to represent the uncertainity
num_uncertainty_qubits = 6

# parameters for considered random distribution

S = 20.0 # intial spot price
vol = 0.6 # volatility of 60%
r = 0.08 # annual interest rate of 8%
T = 30 / 365  # 30 days ( 1 month ) to maturity
````

#### The resulting parameters for log-normal distribution

#### Log Normal Distribution Statistics

**Notation** : $\text{ln } N( \mu, \sigma^2)$

**Probability Distribution** : $\frac{1}{\sqrt{2 \pi \sigma x}} \exp {\frac{(ln(x) - \mu)^2}{2\sigma^2}} $

**Mean** : $e^{\big(\mu + 0.5 \times \sigma^2 \big)}$

**Variance** : $\bigg( e^{\sigma^2} - 1\bigg) \sqrt{(e^{\sigma^2}-1)}$

````python
mu = (r - 0.5 * vol**2) * T + np.log(S)
sigma = vol * np.sqrt(T)
mean = np.exp(mu + sigma**2 / 2)
variance = (np.exp(sigma**2) - 1)* np.exp(2 * mu + sigma**2)
stddev = np.sqrt(variance)
````

#### Defining the bounds



````python
# lowest and highest value for the spot price
low = np.maximum(0,mean- 3* stddev)
high = mean + 3 * stddev
````

````python
# the uncertainity model
uncertainty_model = LogNormalDistribution(num_uncertainty_qubits, mu=mu, sigma = sigma**2, bounds=(low,high))
````

## Plotting the probability distribution

````python
plt.rcParams["figure.figsize"] = (25,10.5)
x = uncertainty_model.values
y = uncertainty_model.probabilities
plt.bar(x,y, width=0.2)
plt.xticks(x,size=15, rotation=90)
plt.yticks(size=15)
plt.grid()
plt.xlabel("Spot Price at Maturity", size = 15)
plt.ylabel("Probability", size=15)
plt.show()
````

### Payoff Function

The payoff function equals to zero as long as the stock price at maturity $S_T$ is less than the strike price $K$ and then increase linearly. The implementation uses a comparator, that flips the ancilla qubit from $|0⟩$ to $|1⟩$ if $S_T \ge K$ and this ancilla is used to control the linear part of the payoff function.

----

The linear part itseld is approximated as follows:

We use the fact that $\sin^2(y+ \pi/4) \approx y + 1/2$

we consider

$$\sin^2( \pi/2 * c_\text{approx} * ( x - 1/2 ) + \pi/4) \approx \pi/2 * c_\text{approx} * ( x - 1/2 ) + 1/2$$

------

We can easily construct an operator that acts as

$$\big|x\rangle \big|0\rangle \mapsto \big|x\rangle \left( \cos(a*x+b) \big|0\rangle + \sin(a*x+b) \big|1\rangle \right),$$

using controlled Y-rotations



````python
# set the strike price

strike_price = 18.96

# approximation scaling of payoff

c_approx = 0.20

````

#### Setting up piecewise linear objective function

````python
breakpoints = [low, strike_price]
slopes = [0,1]
offsets =[0,0]
f_min = 0
f_max = high - strike_price

european_call_objective = LinearAmplitudeFunction(num_uncertainty_qubits,
                                                  slopes,
                                                  offsets,
                                                  domain = (low,high),
                                                  image = (f_min, f_max),
                                                  breakpoints=breakpoints,
                                                  rescaling_factor=c_approx)
````

````python
# compose the model and objective

num_qubits =european_call_objective.num_qubits
european_call = QuantumCircuit(num_qubits)
european_call.append(uncertainty_model, range(num_uncertainty_qubits))
european_call.append(european_call_objective, range(num_qubits))

# draw the circuit

european_call.draw("mpl")
````

#### Plotting and Visualizing


````python
# plot exact payoff function (evaluated on the grid of the uncertainty model)
x = uncertainty_model.values
y = np.maximum(0, x - strike_price)
plt.plot(x, y, "ro-")
plt.grid()
plt.title("Payoff Function", size=15)
plt.xlabel("Spot Price", size=15)
plt.ylabel("Payoff", size=15)
plt.xticks(x, size=15, rotation=90)
plt.yticks(size=15)
plt.show()
````

### Evaluation

````python
# evaluate exact expected value (normalized to the [0, 1] interval)
exact_value = np.dot(uncertainty_model.probabilities, y)
exact_delta = sum(uncertainty_model.probabilities[x >= strike_price])
print("exact expected value:\t%.4f" % exact_value)
print("exact delta value:   \t%.4f" % exact_delta)
````

### Evaluate Expected Payoff

````python
european_call.draw('mpl')
````

````python
# set target precision and confidence level
epsilon = 0.01
alpha = 0.05

qi = QuantumInstance(Aer.get_backend("aer_simulator"), shots=100)
problem = EstimationProblem(
    state_preparation=european_call,
    objective_qubits=[3],
    post_processing=european_call_objective.post_processing,
)
# construct amplitude estimation
ae = IterativeAmplitudeEstimation(epsilon, alpha=alpha, quantum_instance=qi)
````

````python
result = ae.estimate(problem)
````

````python
conf_int = np.array(result.confidence_interval_processed)
print("Exact value:        \t%.4f" % exact_value)
print("Estimated value:    \t%.4f" % (result.estimation_processed))
print("Confidence interval:\t[%.4f, %.4f]" % tuple(conf_int))
````

### Some details about the function used above

#### LogNormalDistribution

```python
class LogNormalDistribution(num_qubits, mu = None, sigma=None, bounds=None, upto_diag=False, name="P(X)")
```


This makes a circuit to encode a discretized log-normal distribution in qubit amplitudes.

-----

A random variable $X$ is log-normal distributed if

$$\log (X) \sim \mathcal{N}(\mu, \sigma^2)$$

The probability distribution function is

$$\mathbb{P}(X = x) = \frac{1}{x\sqrt{2\pi\sigma^2}} e^{-\frac{(\log(x) - \mu)^2}{\sigma^2}}
$$


------


The parameter $σ$ in this class equals the **variance** $\sigma^2$ and not the standard deviation.


This circuit considers the discretized version of $X$ on $2^{\text{num qubits}}$ equidistant points $x_i$ truncated to `bounds`.

The action of this circuit can be written as 

$$ \mathcal{P}_X |0\rangle^n = \sum_{i=0}^{2^n - 1} \sqrt{\mathbb{P}(x_i)} |i\rangle
 $$

 where $n$ is `num_qubits`.




 -----

 The circuit loads the **square root** of the probabilities into the qubit ampllitudes such that the sampling probability, which is the square of the amplitude, equals the probability of the distribution. 


1. **num_qubits** - The number of qubits used to discretize the random variable. For a 1d random variable it is an integer, for **multiple** dimensions a list of integers indicating the number of qubits to use in each dimension.

2. **mu** - The parameter of $\mu$ of the distribution, can be a float or a 1d random variable or a list of floats for a high dimensional random variable.

3. **sigma** - The parameter $\sigma^2$ which is the variance or covariance matrix.


4. **bounds** - The truncation bounds of the distribution as tuples


5. **upto_diag** - Its a `bool` , if True, load the square root of the probabilities up to multiplication with a diagnol for a more efficient circuit.

6. **name** - The name of the circuit.
